## DATA VISUALIZATION

```python
# Reload the dataset since 'Category' was dropped
df_viz = pd.read_csv("/content/Netflix Dataset.csv")

# Split and count genres
genres = df_viz['Category'].str.split(',', expand=True).stack().str.strip().value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=genres.values, y=genres.index, palette='coolwarm')
plt.title('Top 10 Most Common Genres on Netflix')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```

```python
top_genres = genres.head(6)
plt.figure(figsize=(7,7))
plt.pie(top_genres.values, labels=top_genres.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Set2'))
plt.title('Genre Distribution on Netflix')
plt.show()
```

```python
df_viz['Year'] = pd.to_datetime(df_viz['Release_Date'], errors='coerce').dt.year
genre_year = df_viz.assign(Genre=df_viz['Category'].str.split(',')).explode('Genre')
genre_year['Genre'] = genre_year['Genre'].str.strip()
trend = genre_year.groupby(['Year', 'Genre']).size().reset_index(name='Count')

top5_genres = trend.groupby('Genre')['Count'].sum().sort_values(ascending=False).head(5).index
trend_top5 = trend[trend['Genre'].isin(top5_genres)]

plt.figure(figsize=(12,6))
sns.lineplot(data=trend_top5, x='Year', y='Count', hue='Genre', marker='o')
plt.title('Emerging Genre Trends Over the Years')
plt.ylabel('Number of Titles')
plt.show()
```

```python
# Group by Genre and Category (Movie/TV Show) and unstack
genre_category = genre_year.groupby(['Genre', 'Category']).size().unstack(fill_value=0)

# Sort by the count of Movies and select the top 10 genres
genre_category_sorted = genre_category.sort_values('Movie', ascending=False).head(10)

# Plot the results
genre_category_sorted.plot(kind='bar', figsize=(10,6))
plt.title('Top Genres: Movies vs TV Shows')
plt.ylabel('Count')
plt.show()
```

```python
country_genre = genre_year.groupby(['Country', 'Genre']).size().reset_index(name='Count')
top_countries = country_genre.groupby('Country')['Count'].sum().nlargest(5).index
sns.barplot(data=country_genre[country_genre['Country'].isin(top_countries)],
            x='Country', y='Count', hue='Genre')
plt.title('Top Genres by Country')
plt.xticks(rotation=45)
plt.show()
```

```python
country_genre = df_viz.groupby('Country')['Category'].apply(lambda x: x.str.split(',').explode()).reset_index(name='Genre')
country_genre['Genre'] = country_genre['Genre'].str.strip()
country_genre = country_genre.groupby(['Country', 'Genre']).size().reset_index(name='Count')

top_countries = country_genre.groupby('Country')['Count'].sum().nlargest(5).index
sns.barplot(data=country_genre[country_genre['Country'].isin(top_countries)],
            x='Country', y='Count', hue='Genre')
plt.title('Top Genres by Country')
plt.xticks(rotation=45)
plt.show()
```

```python
rating_genre = genre_year.groupby(['Rating', 'Genre']).size().unstack(fill_value=0)
rating_genre.plot(kind='bar', stacked=True, figsize=(12,6))
plt.title('Genre Distribution Across Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
from wordcloud import WordCloud

text = ' '.join(df_viz['Category'].dropna())
wordcloud = WordCloud(width=800, height=400, background_color='black', colormap='Set2').generate(text)

plt.figure(figsize=(10,6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Genres')
plt.show()
```

```python
recent = genre_year[genre_year['Year'] >= 2018]
recent_genres = recent['Genre'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=recent_genres.values, y=recent_genres.index, palette='viridis')
plt.title('Top 10 Emerging Genres (2018–2021)')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```

```python
heat = genre_year.groupby(['Year', 'Genre']).size().unstack(fill_value=0)
plt.figure(figsize=(14,7))
sns.heatmap(heat, cmap='YlOrRd', cbar_kws={'label': 'Number of Titles'})
plt.title('Genre Popularity Heatmap (2008–2021)')
plt.xlabel('Genre')
plt.ylabel('Year')
plt.show()
```
